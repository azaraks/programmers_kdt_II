# 0. Jupyter Notebook 시작하기
**Jupyter Notebook을 사용해 Python 프로그래밍을 진행해봅시다.**

1. Jupyter Notebook의 단축키 익히기  
    - Esc : 명령 모드
        - **Y** : Code Cell
        - **M** : Markdown Cell
        - **A** : 현재 셀 위에 셀 만들기
        - **B** : 현재 셀 아래에 셀 만들기
        - **dd** : 현재 셀 삭제하기
    - Enter : 편집 모드
    - ctrl(cmd) + Enter : 셀 실행하기(Code) / 형식 반영하기(Markdown) 
2. Markdown
    - Header : #, ##, ###  
    - *italic* : _, *  
    - **bold** : __, **
    - ~strikethrough~ : ~
    - unordered list : -, ...
    - ordered list : 1.
